# Exercise 18.2

Evaluate the similarities of the homeworks of the students

tip: https://github.com/orsinium/textdistance

```python
def similarity(data):
    cont2=0
    results = {}
    for row in data['Full_text_clean']:  
        cont=0
        results [cont2]={}
        for row2 in data['Full_text_clean']:
            # Dividir el texto en palabras
            a = set (row.split ()) 
            b = set (row2.split ())
            # Encontrar la interseccion de los conjuntos a y b para determinar las palabras que se repiten en ambos textos
            c = a.intersection(b)
            # Calclar un score de similaridad
            result=float(len(c)) / (len(a) + len(b) - len(c))
            results[cont2][cont]= round(result, 2) 
            cont=cont+1
        cont2=cont2+1   
    
    results=pd.DataFrame(results)    
    assert results.shape == (data.shape[0], data.shape[0])
    
    return results
```
